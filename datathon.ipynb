```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/kiralikbisiklet.csv')
df.head()
```

```python
df.info()
```

```python
#duration sütunu iyileştirme
df['duration'] = df['duration'].apply(lambda x: x.split()[-1])
df.head()
```

```python
df['duration'] = df['duration'].apply(lambda x: float(pd.to_timedelta(x.split()[-1]).seconds // 60) + pd.to_timedelta(x.split()[-1]).seconds % 60 / 100 if pd.notna(x) else None)
df.head()
```

```python
df['rental_start_date'] = pd.to_datetime(df['rental_start_date'])

# 'start_time' sütununu oluştur
df['start_time'] = df['rental_start_date'].dt.hour * 60 + df['rental_start_date'].dt.minute

# Veriyi hazırlanmış veri seti olarak kullan
X = df[['start_time', 'duration']].values
df.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 'rental_start_date' sütununu datetime türüne çevir
df['rental_start_date'] = pd.to_datetime(df['rental_start_date'])

# 'start_hour' sütununu oluştur
df['start_hour'] = df['rental_start_date'].dt.hour

# Kullanmayacağımız sütunları kaldır
df.drop(['rental_end_date'], axis=1, inplace=True)

# Veriyi hazırlanmış veri seti olarak kullan
X = df[['start_hour', 'duration']].values

# K-Means modelini oluştur
kmeans = KMeans(n_clusters=3, random_state=42)

# Modeli eğit
kmeans.fit(X)

# Küme etiketlerini al
labels = kmeans.labels_

# Küme merkezlerini al
centers = kmeans.cluster_centers_

# Kümeleme sonuçlarını görselleştir
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, alpha=0.75)
plt.xlabel('Start Hour')
plt.ylabel('Duration')
plt.title('Clustering Results by Start Hour')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df['rental_start_date'] = pd.to_datetime(df['rental_start_date'])
hourly_avg_duration = df.groupby(df['rental_start_date'].dt.hour)['duration'].mean()

hourly_data = pd.DataFrame({'Hour': hourly_avg_duration.index, 'Ortalama Kullanım Süresi': hourly_avg_duration.values})
hourly_data.sort_values('Hour', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(hourly_data['Hour'], hourly_data['Ortalama Kullanım Süresi'], marker='o')
plt.title('Saatlere Göre Ortalama Kullanım Süresi')
plt.xlabel('Saatler')
plt.ylabel('Ortalama Kullanım Süresi (dakika)')
plt.xticks(range(24))
plt.show()
```

Gün içinde saat bazlı kullanım yoğunluğu

```python
df['rental_start_date'] = pd.to_datetime(df['rental_start_date'])
hourly_total_duration = df.groupby(df['rental_start_date'].dt.hour)['duration'].sum()
hourly_data = pd.DataFrame({'Hour': hourly_total_duration.index, 'Toplam Kullanım Süresi': hourly_total_duration.values})
hourly_data.sort_values('Hour', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(hourly_data['Hour'], hourly_data['Toplam Kullanım Süresi'], marker='o', color='green')
plt.title('Saatlere Göre Toplam Kullanım Süresi')
plt.xlabel('Saatler')
plt.ylabel('Toplam Kullanım Süresi (dakika)')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```
